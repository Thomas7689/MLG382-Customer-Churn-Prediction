```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
# Numerical features
sns.histplot(df['tenure'], kde=True)
plt.title('Distribution of Tenure')
plt.show()

sns.histplot(df['MonthlyCharges'], kde=True)
plt.title('Distribution of Monthly Charges')
plt.show()
```

```python
# Categorical features
#Countplot for Gender
sns.countplot(df['gender'])
plt.title('Distribution of Gender')
plt.show()
```

```python
#Countplot for Churn
sns.countplot(df['Churn'])
plt.title('Distribution of Churn')
plt.show()
```

```python
#Countplot for Contract
sns.countplot(df['Contract'])
plt.title('Distribution of Contract')
plt.show()
```

```python
#Countplot for PaymentMethod
sns.countplot(df['PaymentMethod'])
plt.title('Distribution of PaymentMethod')
plt.show()
```

```python
#Countplot for InternetService
sns.countplot(df['InternetService'])
plt.title('Distribution of InternetService')
plt.show()
```

```python
#Countplot for PhoneService
sns.countplot(df['PhoneService'])
plt.title('Distribution of PhoneService')
plt.show()
```
