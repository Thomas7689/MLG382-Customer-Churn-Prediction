```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
# Churn vs. Contract
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn vs. Contract')
plt.show()
```

```python
# Churn vs. PaymentMethod
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title('Churn vs. Payment Method')
plt.show()
```

```python
# tenure vs. MonthlyCharges
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Tenure vs. Monthly Charges')
plt.show()
```

```python
# Boxplots for numerical values
numerical_features = ['tenure', 'MonthlyCharges']
categorical_features = ['Churn', 'Contract', 'PaymentMethod']

for num_feature in numerical_features:
    for cat_feature in categorical_features:
        plt.figure(figsize=(10,8))
        sns.boxplot(data=df, x=cat_feature, y=num_feature)
        plt.title(f"Boxplot of {num_feature} vs {cat_feature}")
        plt.show()
```
